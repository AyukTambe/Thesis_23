```python
import json
import jsonschema
import pandas as pd

from google.colab import drive
drive.mount('/content/drive')

def find_optional_elements(schema, document_paths):
    # Load the JSON schema
    with open(schema, "r") as schema_file:
        schema_data = json.load(schema_file)

    optional_elements_dict = {}  # Dictionary to store optional elements for each document

    for document_path in document_paths:
        # Load the JSON document
        with open(document_path, "r") as document_file:
            document_data = json.load(document_file)

        # Iterate through the optional properties in the schema
        optional_properties = {}
        for prop, prop_schema in schema_data.get("properties", {}).items():
            if not prop_schema.get("required", False):
                # Check if the optional property is present in the document
                optional_properties[prop] = prop in document_data

        document_name = document_path.split("/")[-1]  # Extract document name
        optional_elements_dict[document_name] = optional_properties

    return optional_elements_dict
# Replace these paths with your actual schema and document paths
schema_path = "/content/drive/My Drive/Thesis_23/DATS_Schema.json"
document_paths = ["/content/drive/My Drive/Thesis_23/36864_dats_2.2-1.json",
                  "/content/drive/My Drive/Thesis_23/3504_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/3545_dats_2.2.json",
                 "/content/drive/My Drive/Thesis_23/4073_dats_2.2.json",
                 "/content/drive/My Drive/Thesis_23/4236_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/32161_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/7219_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/35031_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/37643_dats_2.2.json",
                  "/content/drive/My Drive/Thesis_23/37314_dats_2.2.json"
                  ]

optional_elements_dict = find_optional_elements(schema_path, document_paths)


df = pd.DataFrame(optional_elements_dict).T

# Set pandas options to show all columns
pd.set_option('display.max_columns', None)  # Show all columns
pd.set_option('display.expand_frame_repr', False)  # Don't wrap the DataFrame across multiple lines

# Convert boolean values to check marks (True -> ✓, False -> ❌)
df = df.applymap(lambda x: '✓' if x else '❌')

df
```
